```python
# Load the dataset. Here, I will use only two dataset which I think are helpul for our prediction.
matches = pd.read_csv(r"C:\Users\Harsha\Downloads\Sa project\matches.csv")
deliveries = pd.read_csv(r"C:\Users\Harsha\Downloads\Sa project\deliveries.csv")
```

```python
# checking the shape of both the dataset
print('Matches data shape -->',matches.shape)
print('Deliveries data shape -->',deliveries.shape)
```

```python
# Now we need to do some feature engineering to change the data in such a way we want.
total_score_df = deliveries.groupby(['match_id','inning'])['total_runs'].sum().reset_index()
total_score_df.head()
```

```python
# We will fetch only first inning data because we need to predict a probability in a second inning.
total_score_df_first_inning = total_score_df[total_score_df['inning'] == 1]
total_score_df_first_inning.head()
```

```python
# As we successfully created our first feature, its time to merge the data to the matches dataframe.
matches_df = pd.merge(matches,total_score_df_first_inning[['match_id','total_runs']],left_on='id',right_on='match_id')
matches_df.head()
```

```python
# We will only consider those teams which have played almost all the seasons.
matches_df['team1'].unique()
```

```python
# we can see from the above result that some of the teams have changes their name but basically they are same team so we need to change the name.
matches_df['team1'] = matches_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
matches_df['team2'] = matches_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

matches_df['team1'] = matches_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
matches_df['team2'] = matches_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
ipl_teams = [
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Kolkata Knight Riders',
    'Mumbai Indians',
    'Sunrisers Hyderabad',
    'Royal Challengers Bangalore',
    'Kings XI Punjab',
    'Delhi Capitals'
]
```

```python
matches_df = matches_df[matches_df['team1'].isin(ipl_teams)]
matches_df = matches_df[matches_df['team2'].isin(ipl_teams)]
matches_df.shape
```

```python
# only those matches where dl_applied == 0
matches_df = matches_df.loc[matches_df['dl_applied'] == 0]
```

```python
# Extracting only necessary columns
matches_df = matches_df[['match_id','city','winner','total_runs']]
matches_df.head()
```

```python
# Merge deliveries and matches df for further processing
df_merged = pd.merge(matches_df,deliveries,on='match_id')
df_merged.head()
```

```python
# As we know we will predict a probablity when the second innings will start.
df_merged = df_merged.loc[df_merged['inning'] == 2]
```

```python
# Convert total_runs_y to numeric, forcing non-numeric entries to NaN (if any)
df_merged['total_runs_y'] = pd.to_numeric(df_merged['total_runs_y'], errors='coerce')

# Now apply the cumulative sum group-wise
df_merged['current_score'] = df_merged.groupby('match_id')['total_runs_y'].cumsum()
```

```python
df_merged['runs_left'] = df_merged['total_runs_x'] - df_merged['current_score']
```

```python
# Creating 'ball_left' feature 
ball_per_over = 6
df_merged['balls_left'] = 126 - (df_merged['over'] * ball_per_over + df_merged['ball'])
```

```python
# Replace NaNs with 0s
df_merged['player_dismissed'].fillna(0, inplace=True)

# Convert dismissals to binary (0 if no dismissal, 1 if a player is dismissed)
df_merged['player_dismissed'] = df_merged['player_dismissed'].apply(lambda x: 0 if x == 0 else 1)

# Ensure it's of integer type
df_merged['player_dismissed'] = df_merged['player_dismissed'].astype(int)

# Compute cumulative wickets fallen per match
df_merged['wickets_fallen'] = df_merged.groupby('match_id')['player_dismissed'].cumsum()

# Subtract from total wickets to get wickets left
df_merged['wickets_left'] = 10 - df_merged['wickets_fallen']
```

```python
df_merged.head()
```

```python
# current run rate (crr = runs/overs)
total_balls = 120
df_merged['crr'] = (df_merged['current_score'] * ball_per_over) / (total_balls - df_merged['balls_left'])
```

```python
# required run rate (rrr = (Runs required to win / Balls Remaining  ) x 6)
df_merged['rrr'] = (df_merged['runs_left']*6) / df_merged['balls_left']
```

```python
# match result: 0-lose | 1-win
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

df_merged['result'] = df_merged.apply(result,axis=1)
```

```python
# final df for model building
df = df_merged[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
# just sample the data to remove any bias
df = df.sample(df.shape[0])
```

```python
# drop null values from final df
df.dropna(inplace=True)
```

```python
cities = df['city'].unique().tolist()
print(cities)
```

```python
# drop those rows where balls_left == 0 or negative value
df = df.loc[df['balls_left'] != 0]
df.head()
```

```python
# Separate Dependent and Independent Variables
X = df.drop('result',axis=1)
y = df['result']
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 50)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
