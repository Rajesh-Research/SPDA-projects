```python
# üì¶ Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üéØ Load the datasets the correct way in Google Colab
from google.colab import files
uploaded = files.upload()

# Now, load the uploaded files
sleep_df = pd.read_csv('sleep_health_lifestyle_dataset.csv')  # check exact filename if needed
performance_df = pd.read_csv('Data set.csv')  # check exact filename if needed

# üé® Set Seaborn style for plots
sns.set(style="whitegrid")

# 1Ô∏è‚É£ Sleep Dataset Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=sleep_df,
    x="Sleep Duration (hours)",
    y="Quality of Sleep (scale: 1-10)",
    hue="Stress Level (scale: 1-10)",
    palette="coolwarm",
    alpha=0.7
)
plt.title("Sleep Duration vs Quality of Sleep (colored by Stress Level)")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Quality of Sleep (1-10)")
plt.legend(title="Stress Level", bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()

# 2Ô∏è‚É£ Player Performance Dataset Analysis
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=performance_df,
    x="overall_rating",
    y="match_count",
    hue="gk_attributes_sum",
    palette="viridis",
    alpha=0.7
)
plt.title("Player Overall Rating vs Match Count (colored by GK Attributes Sum)")
plt.xlabel("Overall Rating")
plt.ylabel("Match Count")
plt.legend(title="GK Attributes Sum", bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()

# 3Ô∏è‚É£ Bonus: Correlation Heatmaps
# Sleep dataset correlation
plt.figure(figsize=(8,6))
sleep_corr = sleep_df[['Sleep Duration (hours)', 'Quality of Sleep (scale: 1-10)',
                       'Physical Activity Level (minutes/day)',
                       'Stress Level (scale: 1-10)', 'Daily Steps']].corr()
sns.heatmap(sleep_corr, annot=True, cmap="coolwarm")
plt.title("Sleep Dataset Correlation Matrix")
plt.show()

# Performance dataset correlation
plt.figure(figsize=(8,6))
# Convert relevant columns to numeric, handling errors
performance_df['overall_rating'] = pd.to_numeric(performance_df['overall_rating'], errors='coerce')
performance_df['gk_attributes_sum'] = pd.to_numeric(performance_df['gk_attributes_sum'], errors='coerce')
performance_df['match_count'] = pd.to_numeric(performance_df['match_count'], errors='coerce')
performance_df['league_count'] = pd.to_numeric(performance_df['league_count'], errors='coerce')

# üé® Set Seaborn style for plots
sns.set(style="whitegrid")
plt.title("Player Performance Dataset Correlation Matrix")
plt.show()
```
